```python
import yfinance as yf
import pandas as pd
```

```python
comm_data = pd.read_csv('commodity_futures.csv')
comm_data
```

```python
comm_data.columns.tolist()
```

```python
comm_data = comm_data .drop(columns= ['LEAN HOGS','HRW WHEAT','GASOLINE','LOW SULPHUR GAS OIL','LIVE CATTLE','SOYBEAN OIL','SOYBEAN MEAL',
                            'SOYBEAN MEAL','GASOLINE','SILVER','ULS DIESEL',])
comm_data
```

```python
comm_data.columns.tolist()
```

```python
comm_data = comm_data.set_index('Date')
comm_data
```

```python
comm_data.isnull().sum()
```

```python
comm_data = comm_data.dropna()
```

```python
comm_data.isnull().sum()
```

```python
# Define time range
start_date = '2000-01-01'
end_date = '2023-08-04'
```

```python
tickers = ['^TNX','DX-Y.NYB']
yf_data = pd.DataFrame()
# Loop through each ticker symbol
for ticker in tickers:
    # Download the historical data for the ticker
    data = yf.download(ticker, start=start_date, end=end_date)
    yf_data[ticker] = data['Close']
yf_data = yf_data.rename(columns={'^TNX' : 'BOND_10Y','DX-Y.NYB' : 'Dollar_Index'})
yf_data
```

```python
yf_data.isnull().sum()
```

```python
yf_data = yf_data.reindex(comm_data.index)
yf_data
```

```python
final_df = pd.concat([comm_data,yf_data], axis=1)
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.dropna()
```

```python
final_df.isnull().sum()
```

```python
final_df
```
